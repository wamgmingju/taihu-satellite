```python
# =================================================================
#  太湖 2024年5-7月 MODIS Terra 单星 · 每日 FAI + NDVI
#  运行环境：Google Colab
#  使用方法：
#    1. 改 TAIHU_ASSET_ID 为你的 Asset 路径
#    2. Runtime → Run all
#    3. 授权 Google 账号
#    4. 自动提交全部导出任务，无需手动确认
# =================================================================

# ----- 安装与认证 -----
!pip install earthengine-api geemap -q

import ee
import geemap
import google.colab
from google.colab import auth

auth.authenticate_user()

# =================================================================
# ######################## 参数（改这里） ########################
# =================================================================
TAIHU_ASSET_ID = 'projects/dulcet-nucleus-498308-g8/assets/taihushp'   # ← 改成你的 Asset ID
START_DATE = '2024-05-01'
END_DATE   = '2024-08-01'
SCALE      = 500                              # MODIS 原生分辨率
FOLDER     = 'GEE_Taihu_2024_Summer'          # Google Drive 文件夹名
# =================================================================

# Extract project ID from TAIHU_ASSET_ID
PROJECT_ID = TAIHU_ASSET_ID.split('/')[1]
ee.Initialize(project=PROJECT_ID)
print("✅ GEE 初始化成功")

# ----- 1. 加载太湖边界 -----
taihu = ee.FeatureCollection(TAIHU_ASSET_ID)
print(f"✅ 太湖边界加载成功：{taihu.size().getInfo()} 个要素")

# ----- 2. 加载 MODIS Terra（单星） -----
modis = (ee.ImageCollection('MODIS/061/MOD09GA')
         .filterBounds(taihu)
         .filterDate(START_DATE, END_DATE))

print(f"📡 原始影像数: {modis.size().getInfo()}")

# ----- 3. 去云（state_1km QA 波段） -----
def mask_clouds(image):
    """
    MOD09GA state_1km 波段：
      bit 0-1: cloud state (0=clear, 1=cloudy, 2=mixed)
      bit 8-9: cirrus detected
      bit 10:  internal cloud algorithm flag
    """
    qa = image.select('state_1km')
    mask = (qa.bitwiseAnd(3).eq(0)                           # 仅晴空
            .And(qa.bitwiseAnd(3 << 8).eq(0))                # 无卷云
            .And(qa.bitwiseAnd(1 << 10).eq(0)))              # 内部算法也认为无云
    return image.updateMask(mask).copyProperties(image, ['system:time_start'])

# ----- 4. 波段选取与重命名 -----
def prep_bands(image):
    """
    MOD09GA 关键波段:
      sur_refl_b01: RED    (645 nm)
      sur_refl_b02: NIR    (858.5 nm)
      sur_refl_b06: SWIR1  (1640 nm)
    反射率原始值为 Int16，缩放因子 0.0001，GEE 自动转为浮点。
    """
    red   = image.select('sur_refl_b01').rename('RED')
    nir   = image.select('sur_refl_b02').rename('NIR')
    swir1 = image.select('sur_refl_b06').rename('SWIR1')
    return (image.addBands([red, nir, swir1], None, True)
                .copyProperties(image, ['system:time_start']))

# ----- 5. 计算 FAI + NDVI，内部统一为 Float32 -----
def add_indices(image):
    """
    NDVI = (NIR - RED) / (NIR + RED)
    FAI  = NIR - RED - (SWIR1 - RED) * 0.2146
          系数 0.2146 = (858.5 - 645) / (1640 - 645)
          引自 Hu et al. (2009), Remote Sensing of Environment

    注意：统一 cast 到 Float32，避免导出时类型冲突。
    对 NDVI（范围 [-1,1]）和 FAI（太湖典型 [-200, 800]），
    Float32（7位有效数字）精度完全足够。
    """
    ndvi = (image.normalizedDifference(['NIR', 'RED'])
                .rename('NDVI')
                .toFloat())    # Float64 → Float32，无损

    fai = (image.expression(
               'NIR - RED - (SWIR1 - RED) * 0.2146',
               {'RED':   image.select('RED'),
                'NIR':   image.select('NIR'),
                'SWIR1': image.select('SWIR1')})
              .rename('FAI')
              .toFloat())     # expression 默认 Float32，保证一致也写上

    return image.addBands([ndvi, fai], None, True)

# ----- 6. 日期标签 -----
def add_date_tag(image):
    d = ee.Date(image.get('system:time_start')).format('yyyyMMdd')
    return image.set('date_tag', d)

# ----- 7. 流水线 -----
processed = (modis
             .map(mask_clouds)
             .map(prep_bands)
             .map(add_indices)
             .map(add_date_tag))

# ----- 8. 可用天数统计 -----
dates = sorted(processed.aggregate_array('date_tag').distinct().getInfo())
may   = [d for d in dates if '202405' in d]
jun   = [d for d in dates if '202406' in d]
jul   = [d for d in dates if '202407' in d]

print("=" * 50)
print(f"📊 MODIS Terra 单星 {START_DATE} ~ {END_DATE}")
print(f"   总计可用: {len(dates)} 天（理论最大 92 天）")
print(f"   覆盖率:   {len(dates)/92*100:.1f}%")
print(f"   5月: {len(may)} 天")
print(f"   6月: {len(jun)} 天")
print(f"   7月: {len(jul)} 天")
print(f"   首日: {dates[0]}  末日: {dates[-1]}")
print("=" * 50)

# ----- 9. 预览地图 -----
Map = geemap.Map()
Map.addLayer(taihu, {'color': '#FF0000'}, '太湖边界')

# 取第一天作为预览
first_img = (processed
             .filter(ee.Filter.eq('date_tag', dates[0]))
             .median()
             .clip(taihu))

Map.addLayer(
    first_img.select('FAI'),
    {'min': -200, 'max': 800,
     'palette': ['#1a237e', '#42a5f5', '#ffffff', '#ffeb3b', '#f44336']},
    f'FAI {dates[0]}'
)

Map.addLayer(
    first_img.select('NDVI'),
    {'min': -1, 'max': 1,
     'palette': ['blue', 'white', 'green']},
    f'NDVI {dates[0]}',
    False  # 默认不显示，手动在图层面板打开
)

Map.centerObject(taihu, 10)
Map.addLayerControl()
Map
print("🔍 地图已加载。红色=蓝藻水华，蓝色=清水，绿色=植被。")

# ----- 10. 自动批量导出 ★ -----
print()
print("=" * 50)
print(f"🚀 开始自动导出 {len(dates)} 天...")
print("=" * 50)

for i, date_str in enumerate(dates):
    day_img = (processed
               .filter(ee.Filter.eq('date_tag', date_str))
               .median()
               .clip(taihu)
               .select(['NDVI', 'FAI']))

    task = ee.batch.Export.image.toDrive(
        image=day_img,
        description=f'MODIS_Taihu_{date_str}',
        folder=FOLDER,
        scale=SCALE,
        region=taihu.geometry().bounds(),
        crs='EPSG:4326',
        fileFormat='GeoTIFF',
        maxPixels=1e13
    )
    task.start()   # 自动提交，无需手动确认
    print(f"✅ {date_str}  [{i+1}/{len(dates)}]")

print()
print("=" * 50)
print(f"🎉 全部 {len(dates)} 个任务已自动提交！")
print(f"   查看进度：https://code.earthengine.google.com → Tasks")
print(f"   下载位置：Google Drive → {FOLDER}")
print("=" * 50)
```

```python
# =================================================================
#  太湖任意点 · 2024年5-7月每日 NDVI & FAI（去云修正版）
#  改进：bit0-1 只排除 definite cloud，保留 clear/mixed/not set
#  双星 Terra + Aqua | 单像素精度
# =================================================================

!pip install earthengine-api -q

import ee
import google.colab
from google.colab import auth

auth.authenticate_user()
ee.Initialize()
print("✅ GEE 初始化成功")

# =================================================================
# ######################## 参数 ########################
# =================================================================
TAIHU_ASSET_ID = 'users/你的用户名/taihu'
POINT_LON = 120.1903777
POINT_LAT = 31.2007156
START_DATE = '2024-05-01'
END_DATE   = '2024-08-01'
# =================================================================

point = ee.Geometry.Point([POINT_LON, POINT_LAT])
taihu = ee.FeatureCollection(TAIHU_ASSET_ID)

inside = taihu.filterBounds(point).size().getInfo()
print(f"📍 目标点: ({POINT_LON}, {POINT_LAT})")
print(f"   在太湖范围内: {'是' if inside > 0 else '否'}")

# ----- 1. 双星加载 -----
terra = ee.ImageCollection('MODIS/061/MOD09GA') \
    .filterBounds(point).filterDate(START_DATE, END_DATE) \
    .map(lambda img: img.set('satellite', 'Terra'))
aqua  = ee.ImageCollection('MODIS/061/MYD09GA') \
    .filterBounds(point).filterDate(START_DATE, END_DATE) \
    .map(lambda img: img.set('satellite', 'Aqua'))
modis = terra.merge(aqua)

print(f"📡 Terra: {terra.size().getInfo()} | Aqua: {aqua.size().getInfo()} | 合计: {modis.size().getInfo()}")

# ----- 2. ★ 修正版去云 —— 核心改动 ★ -----
def mask_clouds(image):
    """
    state_1km bit 0-1:
      0 = clear         ✅
      1 = cloudy        ❌ 只排除这个
      2 = mixed         ✅ 保留（太湖夏季大量像元是这个）
      3 = not set       ✅ 保留
    bit 8-9 (cirrus):   不加限制（夏季卷云标定不可靠）
    bit 10 (internal):  不加限制（过于严苛）
    """
    qa = image.select('state_1km')
    mask = qa.bitwiseAnd(3).neq(1)   # ← 唯一的改动：neq(1) 替代 eq(0)
    return image.updateMask(mask).copyProperties(image, ['system:time_start'])

# ----- 3. 波段重命名 -----
def prep_bands(image):
    red   = image.select('sur_refl_b01').rename('RED')
    nir   = image.select('sur_refl_b02').rename('NIR')
    swir1 = image.select('sur_refl_b06').rename('SWIR1')
    return (image.addBands([red, nir, swir1], None, True)
                .copyProperties(image, ['system:time_start']))

# ----- 4. NDVI + FAI -----
def add_indices(image):
    ndvi = (image.normalizedDifference(['NIR', 'RED'])
                .rename('NDVI').toFloat())
    fai = (image.expression(
               'NIR - RED - (SWIR1 - RED) * 0.2146',
               {'RED': image.select('RED'),
                'NIR': image.select('NIR'),
                'SWIR1': image.select('SWIR1')})
              .rename('FAI').toFloat())
    return image.addBands([ndvi, fai], None, True)

# ----- 5. 日期标签 -----
def add_date_tag(image):
    d = ee.Date(image.get('system:time_start')).format('yyyyMMdd')
    return image.set('date_tag', d)

# ----- 6. 流水线 -----
processed = (modis
             .map(mask_clouds)
             .map(prep_bands)
             .map(add_indices)
             .map(add_date_tag))

# ----- 7. 统计 -----
allDates = sorted(processed.aggregate_array('date_tag').distinct().getInfo())
print(f"\n📊 双星覆盖天数: {len(allDates)} / 92")
may_count = len([d for d in allDates if d.startswith('202405')])
jun_count = len([d for d in allDates if d.startswith('202406')])
jul_count = len([d for d in allDates if d.startswith('202407')])
print(f"   5月: {may_count}天 | 6月: {jun_count}天 | 7月: {jul_count}天")

# ----- 8. 逐日单像素提取 -----
results = []

for i, date_str in enumerate(allDates):
    day_imgs = processed.filter(ee.Filter.eq('date_tag', date_str))
    img_list = day_imgs.toList(day_imgs.size())
    n = day_imgs.size().getInfo()

    valid_ndvis = []
    valid_fais  = []

    for j in range(n):
        img = ee.Image(img_list.get(j))
        vals = (img.select(['NDVI', 'FAI'])
                .reduceRegion(
                    reducer=ee.Reducer.first(),
                    geometry=point,
                    scale=500,
                    bestEffort=True,
                    maxPixels=1
                ))

        try:
            ndvi_val = vals.get('NDVI').getInfo()
            fai_val  = vals.get('FAI').getInfo()
            if ndvi_val is not None and fai_val is not None:
                valid_ndvis.append(float(ndvi_val))
                valid_fais.append(float(fai_val))
        except:
            pass    # 该景该点无数据

    if len(valid_ndvis) > 0:
        # 取中值（单景也是中值 = 自身）
        valid_ndvis.sort()
        valid_fais.sort()
        mid = len(valid_ndvis) // 2
        best_ndvi = valid_ndvis[mid]
        best_fai  = valid_fais[mid]
    else:
        best_ndvi = None
        best_fai  = None

    results.append({
        'date': date_str,
        'NDVI': best_ndvi,
        'FAI':  best_fai,
        'valid_shots': len(valid_ndvis)
    })

# ----- 9. 统计 + 判断 -----
valid_days = [r for r in results if r['NDVI'] is not None]
total_days = len(allDates)
print(f"\n✅ 有数据天数: {len(valid_days)} / {total_days} ({len(valid_days)/92*100:.1f}% 覆盖)")
print(f"   无效空白: {total_days - len(valid_days)} 天")

if len(valid_days) > 0:
    median_ndvi = sorted([r['NDVI'] for r in valid_days])[len(valid_days)//2]
    median_fai  = sorted([r['FAI']  for r in valid_days])[len(valid_days)//2]
    print(f"   NDVI 中值: {median_ndvi:.4f}")
    print(f"   FAI 中值:  {median_fai:.0f}")

    ndvi_neg = sum(1 for r in valid_days if r['NDVI'] < 0)
    fai_pos  = sum(1 for r in valid_days if r['FAI']  > 0)
    print(f"   NDVI < 0 (清水): {ndvi_neg}/{len(valid_days)}")
    print(f"   FAI > 0 (蓝藻):  {fai_pos}/{len(valid_days)}")

# ----- 10. 打印表格 -----
print(f"\n{'日期':<12} {'NDVI':>10} {'FAI':>10} {'景数':>4}")
print("-" * 50)
for r in results:
    ndvi_str = f"{r['NDVI']:.4f}" if r['NDVI'] is not None else "N/A"
    fai_str  = f"{r['FAI']:.0f}"  if r['FAI']  is not None else "N/A"
    print(f"{r['date']:<12} {ndvi_str:>10} {fai_str:>10} {r['valid_shots']:>4}")

# ----- 11. 导出 CSV -----
import csv
csv_name = f'taihu_{POINT_LAT}_{POINT_LON}_2024_summer.csv'

with open(csv_name, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['date', 'NDVI', 'FAI', 'valid_shots'])
    for r in results:
        writer.writerow([r['date'], r['NDVI'], r['FAI'], r['valid_shots']])

from google.colab import files
files.download(csv_name)
print(f"\n📥 已下载: {csv_name}")
```
