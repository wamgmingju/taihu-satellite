```python
# =================================================================
#  太湖 2024年5-7月 MODIS Terra 每日 FAI + NDVI
#  NoData 修复完整版
#
#  输出：
#    每天一个 GeoTIFF（Band 1 = NDVI, Band 2 = FAI）
#    太湖外 = NoData（-9999），ArcGIS Pro / QGIS 打开即透明
#
#  运行：Google Colab → Runtime → Run all
# =================================================================

# ----- 安装依赖 -----
!pip install earthengine-api geemap -q

import ee
import geemap
import google.colab
from google.colab import auth

# ----- 认证 -----
auth.authenticate_user()

# IMPORTANT: Replace 'your-google-cloud-project-id' with your actual Google Cloud Project ID.
# Ensure Earth Engine API is enabled for this project.
ee.Initialize(project='your-google-cloud-project-id')
print("✅ GEE 初始化成功")

# =================================================================
# ######################## 参数 ########################
# =================================================================

TAIHU_ASSET_ID = 'projects/dulcet-nucleus-498308-g8/assets/taihushp'      # ← 改成你的 Asset ID
START_DATE     = '2024-05-01'
END_DATE       = '2024-08-01'                   # exclusive，覆盖 5/6/7 月
SCALE          = 500                            # MODIS 原生分辨率（米）
FOLDER         = 'GEE_Taihu_2024_Summer'        # Google Drive 文件夹名
NODATA_VALUE   = -9999                          # NoData 填充值（不跟真实数据冲突）

# 波段说明：
#   Band 1 = NDVI  范围 [-1, 1]
#   Band 2 = FAI   太湖典型范围 [-200, 800]
#   两个指数的有效值都不可能等于 -9999，所以用它做 NoData 标记安全

# FAI 系数来源：
#   Hu et al. (2009), Remote Sensing of Environment
#   k = (λ_nir - λ_red) / (λ_swir1 - λ_red)
#     = (858.5 - 645) / (1640 - 645)
#     = 213.5 / 995 = 0.214572864
FAI_K = 0.2146

# =================================================================
# ######################## 处理流程 ########################
# =================================================================

# ----- 1. 加载太湖边界 -----
taihu = ee.FeatureCollection(TAIHU_ASSET_ID)
taihu_count = taihu.size().getInfo()
print(f"✅ 太湖边界加载成功：{taihu_count} 个要素")

# 拿太湖外包矩形（GEE 导出必须用矩形 region）
taihu_bounds = taihu.geometry().bounds()

# ----- 2. 加载 MODIS Terra 地表反射率 -----
modis_raw = (ee.ImageCollection('MODIS/061/MOD09GA')
             .filterBounds(taihu)
             .filterDate(START_DATE, END_DATE))

print(f"📡 MODIS Terra 原始影像数: {modis_raw.size().getInfo()}")

# ----- 3. 去云函数 -----
def mask_clouds(image):
    """
    MOD09GA 的 state_1km 波段 (QA 质量标记)：
      bit  0-1: cloud state
                0 = clear
                1 = cloudy      ← 排除
                2 = mixed       ← 排除（可能是云边缘/阴影）
                3 = not set, assume clear
      bit  8-9: cirrus detected (1-3 = 有卷云) ← 排除
      bit 10:   internal cloud algorithm flag   ← 排除
    """
    qa = image.select('state_1km')
    mask = (qa.bitwiseAnd(3).eq(0)                           # 仅晴空 (bit0-1 == 0)
            .And(qa.bitwiseAnd(3 << 8).eq(0))                # 无卷云 (bit8-9 == 0)
            .And(qa.bitwiseAnd(1 << 10).eq(0)))              # 内部算法无云
    return image.updateMask(mask).copyProperties(image, ['system:time_start'])


# ----- 4. 波段选取 + 重命名 -----
def prep_bands(image):
    """
    MOD09GA 表面反射率波段：
      sur_refl_b01  → RED    (645 nm,   红波段)
      sur_refl_b02  → NIR    (858.5 nm, 近红外)
      sur_refl_b06  → SWIR1  (1640 nm,  短波红外1)

    原始值 Int16，GEE 自动 × 0.0001 转反射率浮点。
    """
    red   = image.select('sur_refl_b01').rename('RED')
    nir   = image.select('sur_refl_b02').rename('NIR')
    swir1 = image.select('sur_refl_b06').rename('SWIR1')
    return (image.addBands([red, nir, swir1], None, True)
                .copyProperties(image, ['system:time_start']))


# ----- 5. 计算 NDVI + FAI -----
def add_indices(image):
    """
    NDVI = (NIR - RED) / (NIR + RED)
    FAI  = NIR - RED - (SWIR1 - RED) × k

    统一 .toFloat() → Float32，原因：
      - normalizedDifference() 默认输出 Float64
      - expression() 默认输出 Float32
      - GEE 导出要求所有波段同一精度，否则报错：
        "Exported bands must have compatible data types;
         found inconsistent types: Float32 and Float64."
      - Float32 (7 位有效数字) 对 NDVI [-1,1] 和 FAI [-200,800]
        精度完全足够，MODIS 反射率本身噪声 ~0.005 量级。
    """
    ndvi = (image.normalizedDifference(['NIR', 'RED'])
                .rename('NDVI')
                .toFloat())

    fai = (image.expression(
               'NIR - RED - (SWIR1 - RED) * k',
               {
                   'RED':   image.select('RED'),
                   'NIR':   image.select('NIR'),
                   'SWIR1': image.select('SWIR1'),
                   'k':     FAI_K
               })
              .rename('FAI')
              .toFloat())

    return image.addBands([ndvi, fai], None, True)


# ----- 6. 日期标签（按天分组用） -----
def add_date_tag(image):
    d = ee.Date(image.get('system:time_start')).format('yyyyMMdd')
    return image.set('date_tag', d)


# ----- 7. 跑流水线 -----
processed = (modis_raw
             .map(mask_clouds)
             .map(prep_bands)
             .map(add_indices)
             .map(add_date_tag))

# ----- 8. 统计可用天数 -----
all_dates = sorted(processed.aggregate_array('date_tag').distinct().getInfo())

may_dates = [d for d in all_dates if '20240501' <= d <= '20240531']
jun_dates = [d for d in all_dates if '20240601' <= d <= '20240630']
jul_dates = [d for d in all_dates if '20240701' <= d <= '20240731']

print()
print("=" * 55)
print("📊 可用影像统计")
print("-" * 55)
print(f"   时间范围:   {START_DATE} ~ {END_DATE}（共 92 天）")
print(f"   可用天数:   {len(all_dates)} 天")
print(f"   覆盖率:     {len(all_dates)/92*100:.1f}%")
print(f"   5月: {len(may_dates):2d} 天   6月: {len(jun_dates):2d} 天   7月: {len(jul_dates):2d} 天")
print(f"   首日: {all_dates[0]}    末日: {all_dates[-1]}")
print("=" * 55)

# ----- 9. 预览地图 -----
print()
print("🗺️  正在加载预览地图...")

Map = geemap.Map()
Map.addLayer(taihu, {'color': '#FF0000'}, '太湖边界')

# 取第一天做预览
first_date = all_dates[0]
preview_img = (processed
               .filter(ee.Filter.eq('date_tag', first_date))
               .median()
               .clip(taihu))

# FAI 图层 —— 默认显示
Map.addLayer(
    preview_img.select('FAI'),
    {'min': -200, 'max': 800,
     'palette': ['#1a237e', '#42a5f5', '#ffffff', '#ffeb3b', '#f44336']},
    f'FAI - {first_date}',
    True
)

# NDVI 图层 —— 默认隐藏，手动在图层面板打开
Map.addLayer(
    preview_img.select('NDVI'),
    {'min': -1, 'max': 1,
     'palette': ['blue', 'white', 'green']},
    f'NDVI - {first_date}',
    False
)

Map.centerObject(taihu, 10)
Map.addLayerControl()
Map

print(f"   预览日期: {first_date}")
print(f"   红色 = 蓝藻水华 | 蓝色 = 清水 | 绿色 = 植被")
print(f"   如果 FAI 图层看起来正常 → 确认后自动提交导出任务")
print()

# ----- 10. 批量导出（NoData 已修复）-----
print("=" * 55)
print(f"🚀 开始自动提交 {len(all_dates)} 个导出任务")
print("=" * 55)
print()

submitted = 0
skipped   = []

for i, date_str in enumerate(all_dates):
    try:
        # 当天所有影像 median 合成
        day_img = (
            processed
            .filter(ee.Filter.eq('date_tag', date_str))
            .median())

        # ★★★ NoData 修复核心三步 ★★★
        # ① .clip(taihu)    → 把太湖外的像素变为 masked（无效）
        # ② .unmask(-9999)  → 把 masked 像素明确写为 -9999（而非"空"）
        # ③ noDataValue=-9999 → 在 GeoTIFF 文件头写入 NoData 标签
        #
        # 原理：
        #   .clip() 产生的 masked 像素在 GeoTIFF 里没有统一表示方式。
        #   有的软件读成 0，有的读成 NAN，有的直接忽略。
        #   ArcGIS Pro 往往把没有 NoData 标签的 masked 区域当 0 显示，
        #   导致太湖外出现方块。通过 .unmask(-9999) 给它一个明确数值，
        #   再通过 noDataValue 告诉 GeoTIFF「-9999 = 无效」，
        #   ArcGIS / QGIS 打开就会自动透明。
        export_img = (day_img
                      .clip(taihu)                   # ① 太湖外 → masked
                      .unmask(NODATA_VALUE, False)   # ② masked → -9999
                      .select(['NDVI', 'FAI'])
                      .toFloat())                    # ③ 精度统一（双保险）

        task = ee.batch.Export.image.toDrive(
            image=export_img,
            description=f'MODIS_Taihu_{date_str}',
            folder=FOLDER,
            scale=SCALE,
            region=taihu_bounds,
            crs='EPSG:4326',
            fileFormat='GeoTIFF',
            maxPixels=1e13,
            noDataValue=NODATA_VALUE                  # ★ GeoTIFF NoData 标签
        )
        task.start()
        submitted += 1
        print(f"   ✅ {date_str}  [{i+1}/{len(all_dates)}]")

    except Exception as e:
        skipped.append((date_str, str(e)))
        print(f"   ⚠️ {date_str} 提交失败: {e}")

# ----- 11. 汇总 -----
print()
print("=" * 55)
print(f"🎉 导出任务提交完成")
print(f"   成功: {submitted} 个")
if skipped:
    print(f"   失败: {len(skipped)} 个")
    for s in skipped:
        print(f"       {s[0]}: {s[1]}")
print(f"   查看进度: https://code.earthengine.google.com → Tasks 标签")
print(f"   下载位置: Google Drive → {FOLDER}/")
print("=" * 55)
print()
print("📌 ArcGIS Pro 打开后：")
print("   太湖轮廓外 = 自动透明（NoData = -9999）")
print("   如果仍显示矩形方块 → 见下方排查指南")
print()
print("   ArcGIS Pro 手动确认 NoData：")
print("   1. 右键图层 → Symbology")
print("   2. 展开「Background」")
print("   3. 勾选「Display Background Value」")
print("   4. Value 填 -9999，Color 选「No Color」")
print()
print("   QGIS 打开：")
print("   直接拖入 .tif，太湖外自动透明，无需额外设置")
```
